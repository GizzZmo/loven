### 1. Introduksjon (Objekt-orientert + Funksjonell)
Vi kombinerer flere paradigmer:
- **Objekt-orientert**: LovDataClient-klasse for kapsling
- **Funksjonell**: Rene funksjoner med map/filter/reduce
- **Deklarativ**: Pandas for dataanalyse
- **Imperativ**: Logging og feilhåndtering

Kreditering: Data fra Lovdata.no (Stiftelsen Lovdata). Kun offentlig tilgjengelig data brukes.

```python
# Celle 1: Importer og oppsett (selvforklarende)
import requests
import json
import pandas as pd
from pathlib import Path
from typing import Dict, List, Optional
import logging

logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)

# Globale konstanter (deklarativt)
BASE_URL = "https://api.lovdata.no"
PEACE_THEMES = ["energi", "vann", "miljø", "etikk", "selskap", "oppløsning", "fred"]
```

```python
# Celle 2: Objekt-orientert kjerne – LovDataClient (akademisk korrekt)
class LovDataClient:
    """En kapslet klient for Lovdata API – støtter både offentlig søk og lokal data."""
    
    def __init__(self, base_url: str = BASE_URL):
        self.base_url = base_url
        self.session = requests.Session()
        self.session.headers.update({"Accept": "application/json"})
        logger.info("LovDataClient initialisert for Global Peace arbeid.")
    
    def search(self, query: str, limit: int = 20) -> Dict:
        """Funksjonell søkemetode med feilhåndtering."""
        try:
            url = f"{self.base_url}/sok?q={query}&limit={limit}"
            response = self.session.get(url, timeout=10)
            response.raise_for_status()
            data = response.json()
            logger.info(f"Søkte etter '{query}' – fant {len(data.get('hits', []))} treff.")
            return data
        except Exception as e:
            logger.error(f"API-feil: {e}. Bruk lokal backup-data.")
            return {"error": str(e)}
    
    def filter_peace_laws(self, data: Dict) -> List[Dict]:
        """Deklarativ filtrering med list comprehension."""
        hits = data.get("hits", [])
        return [
            hit for hit in hits
            if any(theme in str(hit.get("tittel", "")).lower() for theme in PEACE_THEMES)
        ]
```

```python
# Celle 3: Funksjonell databehandling + Pandas (multi-paradigmatisk)
def analyze_peace_laws(client: LovDataClient, query: str) -> pd.DataFrame:
    """Hovedfunksjon – kombinerer alle paradigmer."""
    raw_data = client.search(query)
    peace_hits = client.filter_peace_laws(raw_data)
    
    # Konverter til Pandas DataFrame (deklarativ analyse)
    df = pd.DataFrame([{
        "tittel": hit.get("tittel"),
        "url": hit.get("url"),
        "relevans_for_fred": True  # allerede filtrert av filter_peace_laws
    } for hit in peace_hits])
    
    print(f"Analysert {len(df)} lover relatert til Global Peace Agreement.")
    return df
```

```python
# Celle 4: Eksempelkjøring for Styrilia 49
client = LovDataClient()

# Eksempler som berører nabolaget og visjonen
queries = [
    "selskapsloven oppløsning",   # Relevant for Jon-Arve 2022
    "vannressursloven",
    "energilov miljø",
    "etikk Oljefondet"
]

for q in queries:
    df = analyze_peace_laws(client, q)
    display(df.head())  # Vis i notebook
```
