### Paradigmer brukt i dette notebooken
- **Objekt-orientert**: `LovDataClient` / `AsyncLovDataClient` for kapsling og gjenbruk
- **Funksjonell**: Rene funksjoner med `map`/`filter`/`reduce`

**Kreditering:** Data fra Lovdata.no (Stiftelsen Lovdata). Kun offentlig tilgjengelig data brukes.

```python
# Celle 1: Importer fra loven-pakken
# Installer foerst: pip install -e '.[notebook]'  (fra rotmappen til prosjektet)

import logging

import nest_asyncio
import pandas as pd

from loven import (
    PEACE_THEMES,
    AsyncLovDataClient,
    LovDataClient,
    analyze_peace_laws,
    batch_analyze,
)

nest_asyncio.apply()  # Nodvendig for asyncio inne i Jupyter

logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)

print('Fredstemaer:', PEACE_THEMES)
```

```python
# Celle 2: Synkron analyse – enkeltspørring
client = LovDataClient()

df = analyze_peace_laws(client, 'energilov miljo', limit=20)
display(df)
```

```python
# Celle 3: Synkron batch-analyse – flere spørringer samlet
queries = [
    'selskapsloven opplosning',
    'vannressursloven',
    'energilov miljo',
    'etikk Oljefondet',
    'klimamal utslipp',
    'boliglov nabolag',
]

df_batch = batch_analyze(client, queries, limit=20)
print(f'Totalt {len(df_batch)} treff fra {len(queries)} sporringer.')
display(df_batch.head(20))
```

```python
# Celle 4: Asynkron batch-analyse – alle spørringer parallelt (raskere)
import asyncio

async_client = AsyncLovDataClient()

peace_queries = [
    'selskapsloven opplosning',
    'vannressursloven energilov',
    'miljo etikk Oljefondet',
    'kraftforbruk vannforsyning',
    'nabolag bolig miljo Eidsvoll',
    'klimamal utslipp',
]

df_async = await async_client.async_peace_batch_analysis(peace_queries, limit=20)
display(df_async.head(12))
```

```python
# Celle 5: Relevansscoring – topp lover sortert etter antall fredstema-treff
if not df_async.empty and 'tema_treff' in df_async.columns:
    top = df_async[df_async['tema_treff'] > 0][['tittel', 'tema_treff', '_query']]
    print('Topp lover etter relevans for fred:')
    display(top)
```

```python
# Celle 6: Eksporter resultater til CSV
output_path = '../results_peace_laws.csv'
df_async.to_csv(output_path, index=False)
print(f'Lagret {len(df_async)} rader til {output_path!r}.')
```

### Konklusjon

Denne multi-paradigmatiske notebooken viser hvordan `loven`-pakken kan brukes til aa:
- **Soke** i Lovdata etter norske lover og forskrifter
- **Filtrere** resultater etter fredstemaer (energi, vann, miljo, etikk, klima, bolig, ...)
- **Analysere** relevans ved aa telle antall tematiske treff
- **Eksportere** resultatene til CSV for videre analyse
